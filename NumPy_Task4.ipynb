```python
import numpy as np
```

```python
iris= np.genfromtxt(r"C:\Users\udayk\Downloads\Iris_clean_-_Copy.csv",dtype= "float16",delimiter=",")
iris
```

### Sepal_length

```python
sepal_length=iris[1:,0:1]  # accessing the sepallength column
sepal_length
```

```python
sepal_length= np.where(sepal_length== -1,np.nan,sepal_length)  # missing values will be replaced by nan
sepal_length
```

```python
med= np.nanmedian(sepal_length)
med
```

```python
sepal_length= np.where(np.isnan(sepal_length),med,sepal_length) # nan will be replaced by median 
sepal_length
```

```python
cond= (sepal_length>10)|(sepal_length<1)
```

```python
sepal_length= np.where(cond,med,sepal_length).astype("int32")
sepal_length                                                 # outliers will be replaced by median
```

**What is the mean value of sepal length**

```python
np.mean(sepal_length)
```

**sort the sepal length in ascending order.**

```python
np.sort(sepal_length)
```

**apply all statistical measures on  sepal length**

```python
np.mean(sepal_length)   #mean
```

```python
np.median(sepal_length)   #median
```

```python
Range= np.max(sepal_length)-np.min(sepal_length)     #Range
Range
```

```python
np.var(sepal_length)   #Variance
```

```python
np.std(sepal_length)     #standard deviation
```

#### IQR

```python
q1= np.percentile(sepal_length,25)
q1
```

```python
q3= np.percentile(sepal_length,75)
q3
```

```python
IQR = q3-q1
IQR
```

```python
lower_limit= q1-1.5*IQR
lower_limit
```

```python
upper_limit= q3+1.5*IQR
upper_limit
```

```python
sepal_length[sepal_length>upper_limit]
```

```python
sepal_length[sepal_length<lower_limit]
```
