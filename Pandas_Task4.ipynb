### 1. Value

```python
df["Position"].fillna("Production Technician I",inplace= True)
```

```python
df.isna().sum()  #all the missing values filled by the values
```

### 2. statastical Value

```python
df= pd.read_excel(r"C:\Users\udayk\Downloads\ethnic_diversity.xlsx")
```

```python
df.head()
```

```python
df.isna().sum()  # The Missing values in each feature/column
```

```python
df["Position"].fillna(df["Position"].mode()[0],inplace= True) #missing values filled by mode

df["State"].fillna(df["State"].mode()[0],inplace= True) # missing  values filled by mode

df["Zip"].fillna(df["Zip"].median(),inplace= True)  #Missing values filled by median

df["Gender"].fillna(df["Gender"].mode()[0],inplace= True)  # missing  values filled by mode

df["MaritalDesc"].fillna(df["MaritalDesc"].mode()[0],inplace= True)

df["CitizenDesc"].fillna(df["CitizenDesc"].mode()[0],inplace= True)

df["Department"].fillna(df["Department"].mode()[0],inplace= True)

df["Salaries"].fillna(df["Salaries"].median(),inplace= True)

df["Race"].fillna(df["Race"].mode()[0],inplace= True)
```

```python
df.isna().sum()
```
